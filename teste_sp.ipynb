```
import pandas as pd
import numpy as np

#!conda install -c conda-forge geopy --yes
import geopy
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
```

```
url = 'https://thiagorodrigo.com.br/artigo/cep-sao-paulo-lista-de-cep-por-bairro-e-cidade-da-grande-sao-paulo/'
web = pd.read_html(url)
df=web[0]
df['CEP'] = '0'
for i in range(len(df['Faixa de CEP'])):
  text = df.loc[i,'Faixa de CEP']
  text = text.split(' ')
  #print(int(text[0]))
  df.loc[i,'CEP'] = '0' + str((int(text[0]) + int(text[2]))//2)
df.loc[21,'CEP'] = '03540'
df.loc[22,'CEP'] = '03640'
df.loc[25,'CEP'] = '03940'
df.loc[48,'CEP'] = '05740'
```

```
# initialize your variable to None
df['Latitude'] = 0
df['Longitude'] = 0

for i, local in enumerate(df['CEP']):
    zona = df.loc[i,'Zona de São Paulo']
    zona = zona.split(' ')
    if zona[2]=='Zona':
      df.loc[i,'Zona de São Paulo'] = zona[3]
    else:
      df.loc[i,'Zona de São Paulo'] = zona[2]
    bairro = df.loc[i,'Bairros Atendidos']
    bairro = bairro.split('-')
    df.loc[i,'Bairros Atendidos'] = bairro[0]
    location = None
    locator = Nominatim(user_agent= 'myGeocoder')
    string = ''
    string = local + ', Sao Paulo, Brazil'
    #print(string)
    while(location is None):
        location = locator.geocode(string)
    df.loc[i,'Latitude'] = location.latitude
    df.loc[i,'Longitude'] = location.longitude

df.head()
```

```
# create map of Sao Paulo using latitude and longitude values
map_sp = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Zona de São Paulo'], df['Bairros Atendidos']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_sp)  
map_sp

```
